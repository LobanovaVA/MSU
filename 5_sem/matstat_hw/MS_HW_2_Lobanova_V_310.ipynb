### Задача 1A Построить оценку методом моментов и методом максимума правдоподобия

#### 1) для параметра распределения Пуассона

Функция вероятности распределения Пуассона: 
$ \quad P(\xi = k) = \frac{\lambda^k}{k!} e^{-\lambda} $

$
    \textit{Метод мометов:} \quad
    EX_i = \lambda \Rightarrow 
    \boxed{ 
        \hat{\lambda} = \overline{X} } 
$


$\textit{Метод максимума правдоподобия:}$

Функция правдоподобия:
$ \quad L(X, \lambda) = P(X_1 = x_1) \cdot\ldots\cdot P(X_n = x_n) $

$$
    L(X, \lambda) = 
    \frac{\lambda^{}x_1}{x_1!} e^{-\lambda} \cdot\ldots\cdot \frac{\lambda^{}x_n}{x_n!} e^{-\lambda} =
    \frac{\lambda^{\overline{X}n}}{x_1! \cdot\ldots\cdot x_n!} e^{-\lambda n}
$$

$$  \ln{L(X, \lambda)} = n\overline{X}\ln{\lambda} - nx - \ln{(x_1! \cdot\ldots\cdot x_n!)}  $$

$$
    \frac{\partial ln{L}}{\partial \lambda} = 
    \left. \frac{n\overline{X}}{\lambda} - n \; \right|_{\hat{\lambda}} = 0 \Rightarrow 
    \frac{n\overline{X}}{\hat{\lambda}} = n \Rightarrow 
    \boxed{ 
        \hat{\lambda} = \overline{X} }    
$$

$$
    \frac{\partial^2 ln{L}}{\partial \lambda^2} =
    \left. -\frac{n \overline{X}}{\lambda^2} 
    \right|_{\,\hat{\lambda}=\overline{X}} = 
    -\frac{n}{\overline{X}} < 0
$$

####  2) для параметра $(x_0, \alpha)$ распределения Парето

Плотность распределения Парето:

\begin{equation*}
    \quad p(x) = \frac{\alpha}{x_0}\left( \frac{x}{x_0}\right)^{-\alpha-1},\,\, x\ge x_0 
    \quad \text{то есть} \quad
        p(x) = 
         \begin{cases}
           \alpha\,x_0^{\alpha}\,x^{-\alpha - 1} & x\ge x_0\\
           0 & x < x_0
         \end{cases}
\end{equation*} 

$\textit{Метод мометов:}$

$$
    E\xi = \int_{-\infty}^{\infty} x p(x) dx = 
    \int_{x_0}^{\infty} \alpha \left( \frac{x_0}{x}\right)^{\alpha} dx = [x = u x_0] = 
    \alpha\,x_0 \int_{1}^{\infty} u^{-\alpha} du = 
    \alpha\,x_0 \left. \frac {u^{1 - \alpha}}{1 - \alpha} \right|_{\,u = 1}^{\,u = \infty} =
    \frac {\alpha\,x_0}{\alpha - 1}
    \; , \quad \alpha > 1
$$

$$
    E\xi^2 = \int_{-\infty}^{\infty} x^2 p(x) dx = 
    \int_{x_0}^{\infty} \alpha x_0 \left( \frac{x_0}{x}\right)^{\alpha - 1} dx = [x = u x_0] = 
    \alpha\,x_0 \int_{1}^{\infty} u^{1 -\alpha} du = 
    \alpha\,x_0^2 \left. \frac {u^{2 - \alpha}}{2 - \alpha} \right|_{\,u = 1}^{\,u = \infty} =
    \frac {\alpha\,x_0^2}{\alpha - 2}
    \; , \quad \alpha > 2
$$

$$
    D\xi = E\xi^2 - (E\xi)^2 = 
    \frac {\alpha\,x_0^2}{\alpha - 2} - \left( \frac {\alpha\,x_0}{\alpha - 1} \right)^2 =
    \frac {\alpha x_0^2}{(\alpha - 2)(\alpha - 1)^2}
    \; , \quad \alpha > 2
$$


$$
    \text{Получили} \quad 
    EX_i = \frac{\alpha x_0}{\alpha - 1} \Rightarrow 
    \overline{X} = \frac{\hat\alpha \hat{x_0}}{\hat\alpha - 1}
    \quad \text{и} \quad 
    DX_i = \frac {\alpha x_0^2}{(\alpha - 2)(\alpha - 1)^2} \Rightarrow 
    \hat{\sigma^2} = \frac {\hat\alpha \hat{x_0}^2}{(\hat\alpha - 2)(\hat\alpha - 1)^2} 
$$

$$
    \text{Тогда} \quad 
    \hat{\alpha}^2 - 2\hat\alpha = \frac{(\overline{X})^2}{\hat{\sigma^2}} \Rightarrow 
    \hat\alpha = 1 \pm \sqrt{1 + \frac{(\overline{X})^2}{\hat{\sigma^2}} } =
    [\text{так как}\,\alpha > 2] = 
    1 + \sqrt{1 + \frac{(\overline{X})^2}{\hat{\sigma^2}} } =
    \frac {\sqrt{\hat{\sigma^2}} + \sqrt{\hat{\sigma^2} + (\overline{X})^2}}{\sqrt{\hat{\sigma^2}}}
$$

$$
    \boxed{ 
        \hat\alpha = 1 + \sqrt{1 + \frac{(\overline{X})^2}{ \frac1{n}\sum\limits_{i=1}^{n}(X_i-\overline X)^2 }}  }
$$

$$
    \hat{x_0} = \frac {(\hat\alpha - 1)\overline{X}}{\hat\alpha} =
    \overline{X} \left( 1 - \frac1{\hat\alpha} \right) =
    \overline{X} \left( 1 - \frac {\sqrt{\hat{\sigma^2}}}
        {\sqrt{\hat{\sigma^2}} + \sqrt{\hat{\sigma^2} + (\overline{X})^2}} \right) =
    \overline{X} \left( \frac {\sqrt{\hat{\sigma^2} + (\overline{X})^2}}
        {\sqrt{\hat{\sigma^2}} + \sqrt{\hat{\sigma^2} + (\overline{X})^2}} \right) =
    \overline{X} \left( 1 + \sqrt{\frac 
            {\hat{\sigma^2}}{\hat{\sigma^2} + (\overline{X})^2}}\,\right)^{-1} =       
$$

$$ =
    \overline{X} \left( 1 + \left( 1 + 
            \frac {(\overline{X})^2}{\hat{\sigma^2}}\right)^{-0,5}\,\right)^{-1}
    \Rightarrow 
    \boxed{
        \hat{x_0} = \overline{X}\, 
        \left( 1 + \left(
        1 + \frac{(\overline{X})^2}{ \frac1{n}\sum\limits_{i=1}^{n}(X_i-\overline X)^2 }
        \right)^{-0,5}\right)^{-1}
        }       
$$


$\textit{Метод максимума правдоподобия:}$

Функция правдоподобия:
$ \quad L(X, \lambda) = p(x_1) \cdot\ldots\cdot p(x_n) $

$$
    L_+(X, x_0, \alpha) = 
    \frac{\alpha\,x_0^{\alpha}}{x_1^{\alpha + 1}} \cdot\ldots\cdot 
    \frac{\alpha\,x_0^{\alpha}}{x_n^{\alpha + 1}} =
    \frac{\alpha^n x_0^{\alpha n}}{(x_1 \cdot\ldots\cdot x_n)^{\alpha + 1}}
    \;, x_1 \ge x_0\,,\ldots, x_n \ge x_0
$$

\begin{equation*} 
    L(X, x_0, \alpha) = 
     \begin{cases}
       L_+(X, x_0, \alpha), & X_{(1)}\ge x_0\\
       0, & \text{иначе}
     \end{cases}
\end{equation*} 

Область, где $L(X, x_0, \alpha) \neq 0$ зависит от парметра $x_0$ - это нерегулярный случай.

Рассмотрим $L_+(X, x_0, \alpha)$ - видно, что максимум будет достигаться при 
$X_{(1)} \Rightarrow \boxed{\hat{x_0} = X_{(1)}}$.

$$  
    \ln{L(X, x_0, \alpha)} = 
    n\ln{\alpha} + \alpha n \ln{x_0} - (\alpha + 1)\ln{(x_1 \cdot\ldots\cdot x_n)}
$$

$$
    \frac{\partial ln{L}}{\partial \alpha} = 
    \left.
        \frac{n}{\alpha} + n\ln{x_0} - \ln{(x_1 \cdot\ldots\cdot x_n)}
    \right|_{\hat\alpha,\hat{x_0}} = 0 \Rightarrow 
    n \left( \frac1{\hat\alpha} + \ln{\hat{x_0}}\right) = 
    \ln{(x_1 \cdot\ldots\cdot x_n)}
    \Rightarrow 
$$

$$
    \Rightarrow  \hat\alpha =
    \left( \frac{\ln{(x_1 \cdot\ldots\cdot x_n)}}{n} - \ln{\hat{x_0}}\right)^{-1} =
    \left( \overline{\ln{(X)}} - \ln{\hat{x_0}}\right)^{-1}
    \Rightarrow 
    \boxed{
        \hat\alpha =
        \left( \overline{\ln{(X)}} - \ln{X_{(1)}}\right)^{-1} }
$$

$$
    \frac{\partial^2 ln{L}}{\partial \alpha^2} = 
    \left. 
        -\frac{n}{\alpha^2}\; 
    \right|_{\hat\alpha,\hat{x_0}} < 0 
$$


### Задача 2D Построить асимптотический доверительный интервал на основе оценки метода моментов

####  1) для вероятности успеха биномиального распределения с известным числом испытаний

Функция вероятности биномиального распределения: 
$ \quad P(\xi = k) = C_n^k\,p^{k}(1-p)^{n-k} $

$$
    EX_i = np \;\Rightarrow\; 
    \overline{X} = n \,\hat{p} \;\Rightarrow\; 
    \hat{p} = \frac{\overline{X}}{n} 
$$

$$
    DX_i = np\,(1-p)\;\Rightarrow\; 
    \hat{\sigma^2} = 
    n\,\frac{\overline{X}}{n} \left(1 - \frac{\overline{X}}{n}\right) =
    \overline{X} \left(1 - \frac{\overline{X}}{n}\right)
$$
	
Мы знаем, что $\overline{X}$ асимптотически нормальная оценка для $EX_i = np$. 

По ЦПТ $ \sqrt{M}\,(\hat{p} - p) \xrightarrow{d} N\,(0,\,\frac{p\,(1-p)}{n}) \;\Rightarrow $

$$
    \Rightarrow\;
    \sqrt{M}\frac{\sqrt{n}\,(\hat{p} - p)}{\sqrt{\hat{p}(1-\hat{p})}} =
    \sqrt{M}\frac{\sqrt{n}\,(\frac{\overline{X}}{n} - p)}
         {\sqrt{\frac{\overline{X}}{n} \left(1 - \frac{\overline{X}}{n}\right)}} =
    \sqrt{M}\frac{\sqrt{n}(\overline{X} - np)}{\sqrt{\overline{X}\,(n - \overline{X})}}
    \xrightarrow{d} N\,(0,\,1) \;\Rightarrow 
$$

$$
    \Rightarrow
    P\,\left( -u_{\frac{1+\gamma}2} \le  
        \frac{\sqrt{Mn}(\overline{X} - np)}{\sqrt{\overline{X}\,(n - \overline{X})}}
        \le u_{\frac{1+\gamma}2} \right) 
        \rightarrow \gamma
    \Rightarrow
    P\,\left( 
        \frac1{n} \left( \overline{X} - u_{\frac{1+\gamma}2}\,
            \frac{\sqrt{\overline{X}\,(n - \overline{X})}}{\sqrt{Mn}}
        \right) \le p \le  
        \frac1{n} \left( \overline{X} + u_{\frac{1+\gamma}2}\,
            \frac{\sqrt{\overline{X}\,(n - \overline{X})}}{\sqrt{Mn}}
    \right) \right)
    \Rightarrow
$$

$$
    \Rightarrow
    \boxed{ \left[ 
        \frac1{n} \left( \overline{X} - u_{\frac{1+\gamma}2}\,
             \frac{\sqrt{\overline{X}\,(n - \overline{X})}}{\sqrt{Mn}} \right) \,;\,
        \frac1{n} \left( \overline{X} + u_{\frac{1+\gamma}2}\,
             \frac{\sqrt{\overline{X}\,(n - \overline{X})}}{\sqrt{Mn}} \right) 
    \right] }
    \;\text{ - ассимптотически доверительный интервал уровня}\;\gamma \quad
$$

#### 2) для вероятности успеха биномиального распределения с неизвестным числом испытаний

$$
    \text{Из метода моментов:} \quad
    \overline{X} = \hat{n}\,\hat{p}\,, \;
    \hat{\sigma^2} = \hat{n}\,\hat{p}\,(1 - \hat{p})  \;\;\Rightarrow\; 
    \hat{p} = 1 - \frac{\hat{\sigma^2}}{\overline{X}} \;\Rightarrow\;

$$

$$

### Задача 3 Практическая часть

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats

pd.set_option ('max_colwidth', 80)
np.random.seed (5)

data = pd.read_csv ("covid_short_data-5.csv")
age_male = np.array ([data['age'][n] 
                     for n in range (len (data['age'])) 
                     if not np.isnan(data['age'][n]) and data['gender'][n] == 'male'])

print ("Estimator of mean: %lf" % age_male.mean())
print ("Estimator of dispersion: %lf" % np.var(age_male))
```

```python
def mean_confidence_interval(sample, confidence_level):
    #a = 1.0 * np.array(sample)
    n = len(sample)
    #se = stats.sem(sample)
    var = np.var(sample)
    mean = sample.mean()
    u = stats.norm.ppf((1 + confidence_level) / 2)
    shift = u * np.sqrt (var / n)
    #shift = u * se
    return mean - shift, mean + shift

interval = mean_confidence_interval(age_male, 0.95)
print ('Confidence interval [%lf,%lf] by confidence level 0.95' % (interval[0], interval[1]))

import statsmodels.stats.api as sms
sms.DescrStatsW(age_male).tconfint_mean() 
```
